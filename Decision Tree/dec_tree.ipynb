```python
import pandas as pd
from pandas import  DataFrame
from sklearn import tree
from sklearn import preprocessing
import warnings
warnings.simplefilter("ignore")

```

```python
data= pd.read_csv('salaries.csv')

data
```

```python
label= preprocessing.LabelEncoder()
data['degree']= label.fit_transform(data['degree'])
data['company']= label.fit_transform(data['company'])
data['job']= label.fit_transform(data['job'])

data
```

```python
reg= tree.DecisionTreeClassifier(criterion='entropy', max_depth= 5)

#training the model

features= DataFrame(data= data, columns=['company', 'job', 'degree'])
target= DataFrame(data= data, columns=['salary_more_then_100k'])

reg.fit(features, target)

reg.score(features, target)
```

```python
#predictions

reg.predict([[2, 1, 0]])
```
