# Multivariable Linear regression #
$$ y= m1 * x1 + m2* x2 +m3 * x3 + b $$ 

```python
# create a data frame for the data in the csv file
import pandas as pd
from sklearn.linear_model import LinearRegression 
from pandas import DataFrame
import numpy as np
```

```python
homedata= pd.read_csv('homeprices.csv')

homedata
```

#### Fill the Null data with median

```python
#check for null values in the data set
#file the NULL data entry with the median
mid= homedata['bedrooms'].median()

homedata['bedrooms']= homedata['bedrooms'].fillna(mid)

homedata
```

```python
features= DataFrame(data= homedata, columns=['area', 'bedrooms', 'age'])
target= DataFrame(data= homedata, columns=['price'])
```

```python
reg= LinearRegression().fit(features, target)

print("Coefficient", reg.coef_)
print("Intercept", reg.intercept_)

#predict for (3000, 3, 40)

print("The prediction for 3000 area, 3 bedroom amnd 40 age is", reg.predict([[3000, 3, 40]])[0][0] )



```

##### Predict salary based on experience , test score and interview score 


```python
hiring_data= pd.read_csv('hiring.csv')

hiring_data
```

```python
#i can see a lot of empty data enteries
#i need to fix them

#1) Solving the test score

test_median= hiring_data['test_score(out of 10)'].median()

#now fill the empty data in test score with test_median

hiring_data['test_score(out of 10)']= hiring_data['test_score(out of 10)'].fillna(test_median)

hiring_data
```

```python
#2) Now solve the error in the experience section
#NaN in experience means no experience
#also need to convert word number to actual number

from word2number import w2n

hiring_data['experience']= hiring_data['experience'].fillna('zero')

hiring_data['experience']= hiring_data['experience'].apply(w2n.word_to_num)

hiring_data
```

```python
#after doing all the changes in the data
#let's create a reg model

f= DataFrame(data= hiring_data, columns= ['experience', 'test_score(out of 10)', 'interview_score(out of 10)']) #features
t= DataFrame(data= hiring_data, columns= ['salary($)'])
```

```python
regression= LinearRegression().fit(f, t)

# y(hat)= m1x1 + m2x2 + m3x3 + b

print('b', regression.intercept_)
print('m', regression.coef_)

print(hiring_data)

regression.predict([[12, 10, 10]])

```
